```python
from ultralytics import YOLO
```

```python
model = YOLO('yolov8n.pt')
```

```python
model.train(
    data='/content/drive/My Drive/project_dataset/dataset/data.yaml',
    epochs=50,
    imgsz=416,
    batch=4,
    project='/content/drive/My Drive/dataset',
    name='yolov8_student_activity',
    workers=2,
    device='cuda',
)
```
